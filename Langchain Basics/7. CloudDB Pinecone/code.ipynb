## Making of a RAG with Pinecone

#### Data Ingestion

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
FILE_PATH = "Ethereum Whitepaper.pdf"
```

```python
loader = PyPDFLoader(FILE_PATH)
```

```python
len(loader.load())
```

```python
pages = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 50
)
```

```python
split_docs = splitter.split_documents(pages)
```

```python
len(split_docs)
```

```python
split_docs
```

```python
# Creating index
index_name = "eth-whitepaper"

if not pc.has_index(index_name):
    pc.create_index(
        name=index_name,
        dimension=768,
        metric="cosine",
        spec=ServerlessSpec(cloud="aws", region="us-east-1")
    )
```

```python
index = pc.Index(index_name)
```

```python
vectorStore = PineconeVectorStore(index=index, embedding=embeddings)
```

```python
uuid_doc = [str(uuid4()) for _ in range(len(split_docs))]
```

```python
uuid_doc
```

```python
retriever = vectorStore.add_documents(documents=split_docs, ids = uuid_doc)
```

```python
retriever = vectorStore.as_retriever()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model = "gemini-1.5-flash")
```

```python
from langchain import hub
prompt = hub.pull("rlm/rag-prompt")
```

```python
import pprint
pprint.pprint(prompt.messages)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
def format_docs(docs):
    return"\n\n".join(doc.page_content for doc in docs)
```

```python
rag_chain = (
    {"context": retriever | format_docs, "question":RunnablePassthrough()}
     | prompt
     | llm
     | StrOutputParser()
)
```

```python
rag_chain.invoke("What is LLM?")
```

```python
rag_chain.invoke("How does Proof of stake works?")
```
